```python
def plot_top_products(order_items, products, top_n=10):
    """
    Plot the top N selling products based on number of units sold.

    Parameters:
    order_items : dataframe containing order item details
    products : dataframe containing product information
    top_n : number of top products to display
    """

    # Count how many times each product was purchased
    product_sales = order_items.groupby("product_id").size().reset_index(name="units_sold")

    # Merge with product names
    product_sales = product_sales.merge(products, on="product_id")

    # Sort products by highest sales
    top_products = product_sales.sort_values(by="units_sold", ascending=False).head(top_n)

    # Plot the bar chart
    plt.figure(figsize=(10,6))
    sns.barplot(data=top_products, x="units_sold", y="product_name")

    plt.title("Top Selling Products")
    plt.xlabel("Units Sold")
    plt.ylabel("Product Name")

    plt.show()
```

```python
def plot_product_revenue(order_items, products, top_n=10):
    """
    Plot the top products based on total revenue generated.
    """

    # Calculate revenue per product
    product_revenue = order_items.groupby("product_id")["price_usd"].sum().reset_index()

    # Merge with product names
    product_revenue = product_revenue.merge(products, on="product_id")

    # Sort by revenue
    top_revenue = product_revenue.sort_values(by="price_usd", ascending=False).head(top_n)

    # Plot
    plt.figure(figsize=(10,6))
    sns.barplot(data=top_revenue, x="price_usd", y="product_name")

    plt.title("Top Revenue Generating Products")
    plt.xlabel("Revenue ($)")
    plt.ylabel("Product Name")

    plt.show()
```

## Revenue by Product

```python
plot_product_revenue(order_items, products)
```

## Order Value Distribution

```python
def plot_order_value_distribution(orders):
    """
    Plot the distribution of order values.
    Helps understand customer spending behavior.
    """

    plt.figure(figsize=(8,5))

    sns.histplot(orders["price_usd"], bins=30, kde=True)

    plt.title("Order Value Distribution")
    plt.xlabel("Order Value ($)")
    plt.ylabel("Frequency")

    plt.show()
```

```python
plot_order_value_distribution(orders)
```

## Repeat vs New Customers

```python
def plot_customer_type_distribution(orders_with_type):
    """
    Plot number of new vs repeat customers.
    """

    # Count customer types
    counts = orders_with_type["user_type"].value_counts()

    plt.figure(figsize=(6,4))

    counts.plot(kind="bar")

    plt.title("New vs Repeat Customers")
    plt.xlabel("Customer Type")
    plt.ylabel("Number of Orders")

    plt.show()
```

```python
# Count number of orders per user
customer_orders = orders.groupby("user_id").size().reset_index(name="order_count")

# Merge order counts with orders dataset
orders_with_type = orders.merge(customer_orders, on="user_id")

# Label customers as New or Repeat
orders_with_type["user_type"] = orders_with_type["order_count"].apply(
    lambda x: "Repeat" if x > 1 else "New"
)
```

```python
plot_customer_type_distribution(orders_with_type)
```

## Monthly Revenue Trend

```python
def plot_monthly_revenue(orders):
    """
    Plot monthly revenue trend to observe business growth.
    """

    # Convert date column to datetime
    orders["created_at"] = pd.to_datetime(orders["created_at"])

    # Extract month
    orders["month"] = orders["created_at"].dt.to_period("M")

    # Calculate monthly revenue
    monthly_revenue = orders.groupby("month")["price_usd"].sum()

    plt.figure(figsize=(10,5))

    monthly_revenue.plot()

    plt.title("Monthly Revenue Trend")
    plt.xlabel("Month")
    plt.ylabel("Revenue ($)")

    plt.show()
```

```python
plot_monthly_revenue(orders)
```
