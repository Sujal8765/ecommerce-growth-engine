```python
import pandas as pd
```

```python
website_sessions = pd.read_csv("data/website_sessions.csv")
orders = pd.read_csv("data/orders.csv")
order_items = pd.read_csv("data/order_items.csv")
order_item_refunds = pd.read_csv("data/order_item_refunds.csv")
website_pageviews = pd.read_csv("data/website_pageviews.csv")
products = pd.read_csv("data/products.csv")
```

```python
website_sessions["created_at"] = pd.to_datetime(website_sessions["created_at"])
orders["created_at"] = pd.to_datetime(orders["created_at"])
website_pageviews["created_at"] = pd.to_datetime(website_pageviews["created_at"])
```

#### Aggregate orders to SESSION level

```python
orders_per_session = (
    orders
    .groupby("website_session_id")
    .agg(
        orders=("order_id", "nunique"),
        revenue=("price_usd", "sum")
    )
    .reset_index()
)

orders_per_session.head()
```

##### Aggregate pageviews to SESSION level

```python
pageviews_per_session = (
    website_pageviews
    .groupby("website_session_id")
    .size()
    .reset_index(name="pageviews")
)

pageviews_per_session.head()
```

```python
session_level = (
    website_sessions
    .merge(orders_per_session, on="website_session_id", how="left")
    .merge(pageviews_per_session, on="website_session_id", how="left")
)

session_level[["orders", "revenue", "pageviews"]] = (
    session_level[["orders", "revenue", "pageviews"]].fillna(0)
)

session_level.head()
```

#### Insight:
Created a session-level dataset where each row represents one website session.

```python
session_level["is_bounce"] = session_level["pageviews"] == 1
```

#### Insight:
Marked sessions with only one pageview as bounced sessions.

```python
conversion_rate = (session_level["orders"] > 0).mean()
total_revenue = session_level["revenue"].sum()
avg_order_value = total_revenue / session_level["orders"].sum()
bounce_rate = session_level["is_bounce"].mean()

conversion_rate, total_revenue, avg_order_value, bounce_rate
```

```python
print(f"Conversion Rate: {conversion_rate:.2%}")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Average Order Value: ${avg_order_value:.2f}")
print(f"Bounce Rate: {bounce_rate:.2%}")
```

```python
channel_summary = (
    session_level
    .groupby(["utm_source","utm_campaign","device_type","utm_content"])
    .agg(
        sessions=("website_session_id", "count"),
        orders=("orders", "sum"),
        revenue=("revenue", "sum"),
        bounce_rate=("is_bounce", "mean")
    )
    .reset_index()
)

channel_summary["conversion_rate"] = (
    channel_summary["orders"] / channel_summary["sessions"]
)

channel_summary["revenue_per_session"] = (
    channel_summary["revenue"] / channel_summary["sessions"]
)

channel_summary
```

### Time-Based Trends (Month-wise & Year-wise)


```python
session_level["month"] = session_level["created_at"].dt.to_period("M")
session_level["year"] = session_level["created_at"].dt.year
```

```python
monthly_trend = (
    session_level
    .groupby("month")
    .agg(
        sessions=("website_session_id", "count"),
        orders=("orders", "sum"),
        revenue=("revenue", "sum")
    )
    .reset_index()
)

monthly_trend
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(monthly_trend["month"].astype(str), monthly_trend["sessions"])
plt.xticks(rotation=45)
plt.title("Monthly Sessions Trend")
plt.xlabel("Month")
plt.ylabel("Sessions")
plt.tight_layout()
plt.show()
```

```python
yearly_trend = (
    session_level
    .groupby("year")
    .agg(
        sessions=("website_session_id", "count"),
        orders=("orders", "sum"),
        revenue=("revenue", "sum")
    )
    .reset_index()
)

yearly_trend
```

```python
plt.figure()
plt.plot(yearly_trend["year"], yearly_trend["revenue"])
plt.title("Yearly Revenue Trend")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
base_path = "data/"

sessions = pd.read_csv(base_path + "website_sessions.csv", parse_dates=["created_at"])
pageviews = pd.read_csv(base_path + "website_pageviews.csv", parse_dates=["created_at"])
orders = pd.read_csv(base_path + "orders.csv", parse_dates=["created_at"])
```

```python
total_sessions = sessions["website_session_id"].nunique()

total_orders = orders["order_id"].nunique()

total_revenue = orders["price_usd"].sum()

conversion_rate = (total_orders / total_sessions) * 100

# Bounce Rate
pv_counts = pageviews.groupby("website_session_id").size().reset_index(name="pageviews")
bounces = pv_counts[pv_counts["pageviews"] == 1]["website_session_id"].nunique()
bounce_rate = (bounces / total_sessions) * 100

avg_rev_per_session = total_revenue / total_sessions

print("Total Sessions:", total_sessions)
print("Bounce Rate:", round(bounce_rate,2))
print("Conversion Rate:", round(conversion_rate,2))
print("Avg Revenue / Session:", round(avg_rev_per_session,2))
```

```python
sessions["month"] = sessions["created_at"].dt.to_period("M").astype(str)
orders["month"] = orders["created_at"].dt.to_period("M").astype(str)

monthly_sessions = sessions.groupby("month").size().reset_index(name="sessions")
monthly_orders = orders.groupby("month").size().reset_index(name="orders")

monthly = monthly_sessions.merge(monthly_orders, on="month", how="left").fillna(0)

fig = go.Figure()

fig.add_trace(go.Scatter(x=monthly["month"], y=monthly["sessions"],
                         mode="lines+markers", name="Traffic"))

fig.add_trace(go.Scatter(x=monthly["month"], y=monthly["orders"],
                         mode="lines+markers", name="Orders"))

fig.update_layout(title="Traffic vs Orders Trend", xaxis_title="Month")

fig.show()
```
